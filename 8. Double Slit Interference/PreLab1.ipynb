```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
plt.style.reload_library()
plt.style.use(['science', 'grid'])
from ipywidgets import interact
```

```python
def InterferencePattern(a,d,lambda_1):
    title = 'a = ' + str(a) + ' mm, d = ' + str(d) + ' mm, $\lambda$ = ' + str(lambda_1) + ' nm' 
    if(lambda_1 == 656):
        color='red'
    elif(lambda_1 == 545):
        color='green'
    else:
        color='black'
    # Defining parameters in the correct units
    a = a*1e-3 # m
    d = d*1e-3 # m
    lambda_1 = lambda_1*1e-9 # m

    # set up a plot window
    fig = plt.figure(figsize=(5, 5), dpi=80)
    ax = fig.add_subplot(1,1,1)

    # set some x and y axes for the plot
    xmin = -1.5
    xmax = 1.5
    ymin = 0
    ymax = 1
    
    # compute an input range of angles (in degrees)
    theta_deg = np.linspace(-1.5,1.5,num=1000) # degrees

    # convert to radians
    theta = np.pi * theta_deg / 180.

    # compute alpha
    alpha = np.pi * a * np.sin(theta) / lambda_1

    # compute beta
    beta = np.pi * d * np.sin(theta) / lambda_1

    # compute the double slit intensity profile
    I_double = np.cos(beta)**2 * (np.sin(alpha) / alpha)**2

    # compute the single slit intensity profile
    I_single = (np.sin(alpha) / alpha)**2

    # make the plot of the curves
    ax.plot(theta_deg,I_double,color=color,label='Double Slit')

    ax.plot(theta_deg,I_single,color='blue',label='Single Slit', linewidth=1)

    # prettify it
    ax.set_xlim(xmin,xmax)
    ax.set_ylim(ymin,ymax)
    ax.set_xlabel(r'$\theta$ (deg)',fontsize=18)
    ax.set_ylabel(r'$I(\theta)/I_{0}$',fontsize=18)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.title(title, fontsize=15)
    # add a legend
    plt.legend(loc='upper left',prop={'size':12})

    #plot the figure
    plt.show()
```

```python
# define parameters given in lab book
a = 0.085 #mm
d = 0.35 #mm
lambda_laser = 656 # nm
lambda_light = 545  # nm
InterferencePattern(a,d,lambda_laser)
```

```python
interact(InterferencePattern, a=(0.085,0.085,0.005), d=(0.35,0.35,0.05), lambda_1=(545,656,1))
# a: Slit width
# d: Spacing between slits
```
