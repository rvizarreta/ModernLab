```python
#import numpy to handle arrays
import numpy as np

#import matplotlib to create plots
import matplotlib.pyplot as plt

#allows for curve fitting
import scipy.optimize as opt

#import data

volt,unc_volt,d1_avg,d2_avg,d3_avg,d4_avg,unc_d = np.loadtxt('elec_diff.txt' ,unpack=True)
lambda_debroglie=np.array([0.0274,0.0245,0.0224,0.0207,0.0194,0.0183,0.0174]) #IN NANOMETERS

#the function we are fitting, called func
def func1(x,a,b):
    return a*x+b
def func2(x,a):
    return a*x

#fit the data
#The sigma = means you will be doing a fit weighted by the uncertainty in y

#fit WITH a y-intercept
parameters1, covariance1 = opt.curve_fit(func1, lambda_debroglie, d1_avg)#sigma = yerror)
parameters2, covariance2 = opt.curve_fit(func1, lambda_debroglie, d2_avg)#sigma = yerror)
parameters3, covariance3 = opt.curve_fit(func1, lambda_debroglie, d3_avg)#sigma = yerror)
parameters4, covariance4 = opt.curve_fit(func1, lambda_debroglie, d4_avg)#sigma = yerror)

#fit WITHOUT a y-intercept
parameters5, covariance5 = opt.curve_fit(func2, lambda_debroglie, d1_avg)#sigma = yerror)
parameters6, covariance6 = opt.curve_fit(func2, lambda_debroglie, d2_avg)#sigma = yerror)
parameters7, covariance7 = opt.curve_fit(func2, lambda_debroglie, d3_avg)#sigma = yerror)
parameters8, covariance8 = opt.curve_fit(func2, lambda_debroglie, d4_avg)#sigma = yerror)

#Plots the data.  You can play around with the paramters of the plot to change color, shape, etc
#You will likely want to give it a better name than 'data' and will want to change colors, etc
# when you have multiple fits ont he same page
```

```python
plt.figure(figsize=((6,5)))

plt.errorbar(lambda_debroglie,d1_avg,capsize = 5,marker = 'o',linestyle = 'None', label = 'data') #anna throw in y error
#plt.plot(lambda_debroglie,func1(lambda_debroglie,*parameters1),label = 'fit with y-intercept')
plt.plot(lambda_debroglie,func2(lambda_debroglie,*parameters5),label = 'fit without y-intercept')

#insert the legend
plt.legend(loc = 'upper left')


#This prints the fitting parameter
print(parameters1)
print(parameters5)

#calculate the error on the fits from the covariance matrix
perr = np.sqrt(np.diag(covariance1))

#Print the erros on the fits

print(perr)

#axes titles

plt.xlabel('de Broglie wavelength (nm)', fontsize = 16)
plt.ylabel('Bragg wavelength (nm)', fontsize = 16)
plt.title('Smaller inner ring', fontsize = 18)


#Show the plot!
plt.show()


# In[6]:

```
